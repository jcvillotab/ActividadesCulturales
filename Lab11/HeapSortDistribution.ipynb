# Heap Sort Algorithm
---

## Explanation:

The Heap Sort algorithm consists of two key parts, the first is the data structure known as the Heap and the second is a function called *heapify* which transforms a list into a Heap.

### What's a Heap?

A Heap is a data structure based on the Binary Tree which fulfils the property that the root is greater than the children in all its branches or vice versa (Being a min heap and a max heap respectively). 

![Max Heap](https://drive.google.com/uc?id=1DRdoxZcRrkco1FwjWRI1BTO93N8xWkiG)

<br>

### What does the Heapify function do?

The heapify function aims to convert, as mentioned above, a binary tree into a heap by following the instructions below:

First, the root node value is taken and checked to see if it is smaller than the children's value. If this is true, the node is swapped with the child until this condition is false. This is done recursively until the heap is fully converted. Also, this function assumes that all the children trees are Heaps.

![Swap process](https://drive.google.com/uc?id=1YcPRJNFlz58AtvOn5fY2hZQPK6mQzyw9)

<br>

### How does the Heap Sort work?

By having a Max Heap, we ensure that the largest element is at the root of the Heap, so we put it at the end of the array and then its swaped with the bottom rightmost node of the tree. Following this, we call the heapify function which returns the next largest element to the root. The process is repeated until there are no more elements in the heap.

This leads to having the array sorted after $n â‹… log(n)$ steps.

![Heap Sort](https://drive.google.com/uc?id=19MIT2caAm1eeuVvd34555WtYLBVF1PhT)

And that's how the Heap Sort algorithm works.


```python
#Taken from https://www.geeksforgeeks.org/python-program-for-heap-sort/
def heapify(arr, n, i):
    largest = i  # Initialize largest as root
    l = 2 * i + 1     # left = 2*i + 1
    r = 2 * i + 2     # right = 2*i + 2
  
    # See if left child of root exists and is
    # greater than root
    if l < n and arr[i] < arr[l]:
        largest = l
  
    # See if right child of root exists and is
    # greater than root
    if r < n and arr[largest] < arr[r]:
        largest = r
  
    # Change root, if needed
    if largest != i:
        arr[i],arr[largest] = arr[largest],arr[i]  # swap
  
        # Heapify the root.
        heapify(arr, n, largest)

# The main function to sort an array of given size
def heapSort(arr):
    n = len(arr)
  
    # Build a maxheap.
    # Since last parent will be at ((n//2)-1) we can start at that location.
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)
  
    # One by one extract elements
    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]   # swap
        heapify(arr, i, 0)
```

# Heap Sort time complexity histograms for permutations of $n$ different elements counting steps

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.arange(1,21,1)
y = np.log(n)

plt.plot(n, n*y, label = 'Heap sort best time $(n \cdot log n)$')
plt.plot(n, n*y , label = 'Heap average time $(n \cdot log n)$')
plt.plot(n, n*y, label = 'Heap worst time $(n \cdot log n)$')
plt.xlabel('Array size')
plt.ylabel('Steps')
plt.legend(loc = 'upper left')
```

```python
def heapifySteps(arr, n, i):
    counter = 0
    largest = i 
    l = 2 * i + 1     
    r = 2 * i + 2
    counter += 3     
  
    if l < n and arr[i] < arr[l]:
        largest = l
        counter += 1
    
    if r < n and arr[largest] < arr[r]:
        largest = r
        counter += 1
    counter += 2
    if largest != i:
        arr[i],arr[largest] = arr[largest],arr[i]  
        counter += heapifySteps(arr, n, largest) 
    #print("Inside: " , counter)
    return counter

def heapSortSteps(arr):
    steps = 0
    n = len(arr)
  

    for i in range(n // 2 - 1, -1, -1):
        steps += heapifySteps(arr, n, i) +1
        #print("Outside 1:" , steps)
  

    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        steps += heapifySteps(arr, i, 0) +3
        #print("Outside 2:" , steps)
    return steps
```

```python
#Taken from https://www.geeksforgeeks.org/iterative-heap-sort/#_=_
def buildMaxHeap(arr, n):
	counter = 0
    
	for i in range(n):
		counter +=1
		# if child is bigger than parent
		if arr[i] > arr[int((i - 1) / 2)]:
			counter += 1
			j = i
            
		
		
			while arr[j] > arr[int((j - 1) / 2)]:
				(arr[j],
				arr[int((j - 1) / 2)]) = (arr[int((j - 1) / 2)],arr[j])
				j = int((j - 1) / 2)
				counter +=3
			counter += 1
	return counter
def heapSortIteSteps(arr, n):
	count = 0
	count += buildMaxHeap(arr, n)

	for i in range(n - 1, 0, -1):
		
		# swap value of first indexed
		# with last indexed
		arr[0], arr[i] = arr[i], arr[0]
		count += 3
        
	
		# maintaining heap property
		# after each swapping
		j, index = 0, 0
		count += 2
		
		while True:
			index = 2 * j + 1
			count += 2
			
			# if left child is smaller than
			# right child point index variable
			# to right child
			if (index < (i - 1) and
				arr[index] < arr[index + 1]):
				index += 1
				count +=1
			count +=1
		
			# if parent is smaller than child
			# then swapping parent with child
			# having higher value
			if index < i and arr[j] < arr[index]:
				arr[j], arr[index] = arr[index], arr[j]
				count +=2
            
		
			j = index
			count +=2
			if index >= i:
				count +=1
				break
	return count
```

Generate all permutations without repetitions of size n (for small n), plot the "exact" time distribution (number of steps), and print the media and the standard deviation

```python
from itertools import permutations
import matplotlib.pyplot as plt
import numpy as np
import math

n=3
X = range(n)
times = []

for P in permutations(X):
  t = heapSortIteSteps(list(P), len(P)) 
  times.append(t)  
b = np.array(range(min(times)-1, max(times)+1))+0.5
plt.hist(times, bins = b )
plt.title("Heap Sort")
plt.xlabel("Steps")
plt.ylabel("Frequency")
plt.show()
    
print("n",n)
print("number of permutations n! =",math.factorial(n)) 
print("exp min =",min(times))
print("min formula n* log n =",n*np.log(n))
print("exp max =",max(times))
print("max formula n*log n =",n*np.log(n))
print("exp mean =",np.mean(times))
print("mean formula n*log n =",n*np.log(n))
print("stdev",np.std(times))
```

```python
# https://towardsdatascience.com/an-easy-introduction-to-3d-plotting-with-matplotlib-801561999725
from itertools import product
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import random 

fig = plt.figure()
ax = plt.axes(projection="3d")

N = 8

x = []
y = []
z = []
dx = []
dy = []
dz = []

for n in np.arange(3,N+1,1):

   times = []

   for P in product(range(1,n+1), repeat = n):
       t = heapSortIteSteps(list(P), len(P))
       times.append(t)  

   b = np.array(range(min(times)-1, max(times)+1))+0.5
   H = np.array(np.histogram(times,b))
   H[0] = H[0]/np.sum(H[0])
   for i in range(len(H[0])):
     x.append(n)
     y.append(H[1][i])
     z.append(0)
     dx.append(0.1)
     dy.append(0.1)
     dz.append(H[0][i])

ax.bar3d(x, y, z, dx, dy, dz, color='aqua', zsort='average')
plt.show()
```
